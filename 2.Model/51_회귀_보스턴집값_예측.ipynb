# 선형회귀-보스턴집값예측

```
import numpy as np
import tensorflow as tf
```

```
# seed 값 설정
seed = 2021
np.random.seed(seed)
tf.random.set_seed(seed)
```

```
from sklearn.datasets import load_boston
boston = load_boston()
```

```
boston.data.shape, boston.target.shape
#((506, 13), (506,))
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, random_state=seed #test_size=0.2디폴트
)

X_train.shape, y_train.shape
#((379, 13), (379,))
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

### model = Sequential시, Output Layer의 activation함수 없음

```
model = Sequential([
    Dense(30, input_dim=13, activation='relu'),
    Dense(6, activation='relu'),
    Dense(1) #회귀라서 그냥 아웃풋
])
model.summary()
```

### loss함수는mean_squared_erro

```
model.compile(loss='mean_squared_error', optimizer='adam') #회귀의 loss함수는mean_squared_error 에러를제곱한것의평균
```

```
model.fit(X_train, y_train, epochs=500, batch_size=20)
```

### 회귀평가

```
y_prediction = model.predict(X_test)
y_prediction.shape 
#(127, 1) 2차원[[],[],[],[]]이걸 이렇게[ , , , ]차원줄이려고 flatten()펼쳐주세욧~한차원내려줌
```

```
# 예측 값과 실제 값의 비교
y_prediction = model.predict(X_test).flatten()
for i in range(10): #y_test의 10개정도 
    label = y_test[i]
    prediction = y_prediction[i]
    print(f"실제가격: {label:.3f}, 예상가격: {prediction:.3f}")
#실제가격: 21.700, 예상가격: 18.637
#실제가격: 15.600, 예상가격: 15.172
#실제가격: 20.000, 예상가격: 19.492
#실제가격: 12.800, 예상가격: 11.534
```
